```python
def get_time_signature(num_beats_per_measure, num_quavers_per_beat):
    if num_quavers_per_beat == 2:
        if num_beats_per_measure == 2:
            return '4/4'
        if num_beats_per_measure == 3:
            return '3/4'
    if num_quavers_per_beat == 3:
        if num_beats_per_measure == 2:
            return '6/8'
        if num_beats_per_measure == 3:
            return '9/8'
    
    print(num_beats_per_measure, num_quavers_per_beat)
    return '?'
```

```python
beats_per_measures = index_df['num_beats_per_measure'].tolist()
quavers_per_beats = index_df['num_quavers_per_beat'].tolist()

time_signatures = [get_time_signature(
    beats_per_measures[i], quavers_per_beats[i]) for i in range(len(index_df))]
```

# Phrase & Section

```python
index_df = index_df[['song_id', 'name', 'artist', 'time_signature']]
index_df
```

```python
def get_pretty_structure(phrases):
    sections = []

    curr = 0
    curr_sep = []
    curr_section = []

    while curr <= len(phrases) - 1:
        melody_type = phrases[curr]
        curr += 1

        measures = ''
        while curr <= len(phrases) - 1 and phrases[curr].isdigit():
            measures += phrases[curr]
            curr += 1
        measures = int(measures)

        word = '{}{}'.format(melody_type, measures)

        if melody_type.islower() or (melody_type == 'X' and measures > 2):
            if len(curr_section) != 0:
                sections.append(' '.join(curr_section))
                curr_section = []

            curr_sep.append(word)

        else:
            if len(curr_sep) != 0:
                sections.append(' '.join(curr_sep))
                curr_sep = []

            curr_section.append(word)

    if len(curr_section) == 0 and len(curr_sep) == 0:
        print('ERROR')
        return None

    if len(curr_section) != 0:
        sections.append(' '.join(curr_section))

    if len(curr_sep) != 0:
        sections.append(' '.join(curr_sep))

    raw_len = len(phrases)
    new_len = len(' '.join(sections).replace(' ', ''))
    assert raw_len == new_len

    return get_pretty_section(sections)


def get_pretty_section(sections):
    pretty_sections = []

    num_measures = 1
    for sec in sections:
        pretty_sec = []
        for word in sec.split():
            tmp_num = int(word[1:])
            pretty_sec.append('{}~{}: {}'.format(
                num_measures, num_measures + tmp_num - 1, word))
            num_measures += tmp_num
        pretty_sections.append('; '.join(pretty_sec))

    return pretty_sections


def get_structure_labels(song_id):
    structure_dir = '../../MusicData/hierarchical-structure-analysis/POP909/{:03d}'.format(
        song_id)
    with open(os.path.join(structure_dir, 'human_label1.txt'), 'r') as f:
        label1 = f.readlines()[0].strip()
    with open(os.path.join(structure_dir, 'human_label2.txt'), 'r') as f:
        label2 = f.readlines()[0].strip()

    sections_1 = get_pretty_structure(label1)
    sections_2 = get_pretty_structure(label2)

    return sections_1, sections_2
```

```python
structure_labels = [get_structure_labels(
    song_id) for song_id in tqdm(index_df['song_id'])]
```

```python
def double_elem_of_dataframe(col_name):
    elems = index_df[col_name].tolist()
    elems = [(e, e) for e in elems]
    elems = [a for b in elems for a in b]
    return elems
```

```python
song_ids = double_elem_of_dataframe('song_id')
names = double_elem_of_dataframe('name')
artists = double_elem_of_dataframe('artist')
time_signatures = double_elem_of_dataframe('time_signature')
structures = [' / '.join(a) for b in structure_labels for a in b]

df = pd.DataFrame(dict(zip(['song_id', 'name', 'artist', 'time_signature', 'structure'], [
                  song_ids, names, artists, time_signatures, structures])))
len(index_df), len(df)
```

# 小节数量查看

```python
from partition import *
import muspy
import pandas as pd
from tqdm import tqdm
import random
```

```python
index = pd.read_csv('./partition/index.csv')
index
```

```python
res = []
for song_id in tqdm(index[index['time_signature'] == '4/4']['song_id']):
    midi, _, bar_window, _ = read_midi(song_id)
    
    l1, l2 = get_structure_labels(song_id)
    phrases_1 = [p for sec in l1 for p in sec]
    phrases_2 = [p for sec in l2 for p in sec]
    
    num_bars = midi.get_end_time() // bar_window + 1
    label_bars_1 = sum(p['duration'] for p in phrases_1)
    label_bars_2 = sum(p['duration'] for p in phrases_2)
    
    
    res.append((song_id, num_bars, label_bars_1, label_bars_2))
```

```python
for i, num_bars, label_bars_1, label_bars_2 in res:
    if label_bars_1 != label_bars_2:
        print(i, num_bars, label_bars_1, label_bars_2)
```

```python
for i, num_bars, label_bars_1, label_bars_2 in res:
    if label_bars_1 != label_bars_2 and abs(label_bars_1 - label_bars_2) != 1:
        print(i, num_bars, label_bars_1, label_bars_2)
```
